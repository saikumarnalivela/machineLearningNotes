# Plot the log-loss function
Given and instance, the model will produce a probability p between 0 and 1 that indicates the prediction that the instance is positive.  y is a binary indicator for the instance (0=negative, 1=positive)

```python
import math
def log_loss(p,y):
    return -(y*math.log(p) + (1-y)*math.log(1-p))
log_loss(.9, 0)

#math.log(.1)
```

```python
points = 1000
delta = 1.0/points
p_data = [i*delta for i in range(1,points)]
ll_data = [log_loss(p, 0) for p in p_data]

import matplotlib.pyplot as plt
plt.plot(p_data, ll_data)
```
