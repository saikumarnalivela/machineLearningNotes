It brings up an important point.  How do you know if a sequence of values is random or not?  I might be able to look into how the values were generated and come up with a physical reason for why they are random, but it is a deep question.  If I just look at a sequence of digits from 0 to 9 every sequence is equally likely. That means the sequence 9999999999 is just as likely as 2193347198.  And yet, we would be less likely to accept 9999999999 as random.

A couple basic points:
*Humans are really, really likely to detect patterns in random data. This shows up all the time.  For example, people believe that streaks have meaning.  There is a common belief in basketball that a shooter can get into a hot streak. Which leads to coach instructions to get the ball to the player with the hot hand.
*Humans are really, really bad at probability and statistics. Can we prove that a sequence is random?  At best, we can collect statistics about the sequence and compute a probability that a given sequence would occur.

Example:
If I flip a fair coin 20 times and it comes up heads each time, which way should I bet
1. Bet a lot that the next coin flip will be a head, because past history predicts that we have another head coming up.
2. Don't bet anything. It's a fifty-fifty shot.
3. Bet a lot that the next coin flip will be a tail, because we are due to get a tail

The correct answer is 2. It's a fair coin and past behavior does not indicate future result.  

_The even more correct answer._ This is not a fair coin, so go with heads.  (There is only a 1/1000000 chance that all heads comes up.  

_The even more correct correct answer._ This is not a fair coin, and the game is rigged.  If I bet anything, I will lose.

Case 1 is the hot hand theory.  Who should I have shoot the basketball at the end of the game?  Who should I give money to invest in the stock market.  In both cases, there is good evidence that recent past performance does not predict future results. 

Case 3 is the gambler's fallacy.

Run the following code.  Look at the two plots and then decide which one is less random.


```python
from random import uniform

x1 = [uniform(0.0,100.0) for x in range(100)]
y1 = [uniform(0.0,100.0) for x in range(100)]

x2 = [uniform(0.0,10.0) + (x//10)*10.0 for x in range(100)]
y2 = [uniform(0.0,10.0) + (x%10)*10.0for x in range(100)]


import matplotlib.pyplot as plt



plt.plot(x1,y1, '.')
plt.show()

plt.plot(x2,y2, '.')
plt.show()
```

Most people will look at the first plot and see patterns and decide that it is less random. In fact, the dots are spread out evenly across the second plot and are less random.  We have one point in each of the 100 grid boxes.

```python
from random import uniform

x1 = [uniform(0.0,100.0) for x in range(100)]
y1 = [uniform(0.0,100.0) for x in range(100)]

x2 = [uniform(0.0,10.0) + (x//10)*10.0 for x in range(100)]
y2 = [uniform(0.0,10.0) + (x%10)*10.0for x in range(100)]


import matplotlib.pyplot as plt


def add_grid(plot):
    for x_vert in range(11) :
        vert_bar_x = []
        vert_bar_x.append(x_vert*10)
        vert_bar_x.append(x_vert*10)
        vert_bar_y = [0, 100]
        plt.plot(vert_bar_x, vert_bar_y)
    
    for y_hor in range(11) :
        hor_bar_x = [0, 100]
        hor_bar_y = []
        hor_bar_y.append(y_hor*10)
        hor_bar_y.append(y_hor*10)
        plt.plot(hor_bar_x, hor_bar_y)


plt.plot(x1,y1, '.')
add_grid(plt)
plt.show()

plt.plot(x2,y2, '.')
add_grid(plt)
plt.show()
```
